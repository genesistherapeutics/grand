```python
import mdtraj as md
import os
import glob
```

```python
traj = md.load('proposal_tracking.pdb')
```

```python
traj_list = []
f = open('ghosts_in_sys.txt','r')
for i, line in enumerate(f.readlines()):
    selection_line = ""
    for res_id in line.strip().split(','):
        selection_line += f" or (resi {res_id})"
    traj_list.append(traj[i].atom_slice(traj.top.select(f'(resn != UNK) {selection_line}')))
```

```python
traj_list[-1].top
```

```python
for i, traj in enumerate(traj_list):
    traj.save_pdb(f'output/real_only_{i}.pdb')
```
