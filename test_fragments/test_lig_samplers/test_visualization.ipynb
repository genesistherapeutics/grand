```python
import mdtraj as md
import numpy as np
import openmm
import openmm.app
```

```python
traj = md.load('phenol_sacp.dcd',top='gcmc-ghosts.pdb')
```

```python
pdb = openmm.app.PDBFile('gcmc-ghosts.pdb')
```

```python
prot_atoms = []
prot_residue = []
for residue in pdb.topology.residues():
    if residue.name != 'UNK':
        prot_residue.append(residue)
        for atom in residue.atoms():
            prot_atoms.append(atom.index)
```

```python
prot_residue
```

```python
anchor_molecules=[set(traj.topology.residue(0).atoms)]
```

```python
traj.superpose(traj[0],atom_indices=np.array(prot_atoms))
traj = traj.image_molecules(anchor_molecules=anchor_molecules)
```

```python
traj.save_dcd('superposed.dcd')
```
